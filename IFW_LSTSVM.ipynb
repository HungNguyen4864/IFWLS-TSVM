```python
import numpy as np
import time
from non_linear_score_values import NonLinearScoreValues
from linear_W_interclass_weights import LinearWInterclassWeights

def IFW_LSTSVM(TestX, DataTrain, FunPara):
    obs = DataTrain[:, -1]
    A = DataTrain[obs == 1, :-1]
    B = DataTrain[obs != 1, :-1]

    c1 = FunPara['c1']
    c2 = FunPara['c2']
    epsilon = FunPara['c3']
    K = FunPara['k']
    kerfPara = FunPara['kerfPara']

    m1 = A.shape[0]
    m2 = B.shape[0]
    e1 = np.ones((m1, 1))
    e2 = np.ones((m2, 1))

    start_time = time.time()

    nlsv = NonLinearScoreValues(DataTrain, kerfPara['pars'])
    S1, S2, alpha_d, mew = nlsv.compute_scores()

    liw = LinearWInterclassWeights(DataTrain, kerfPara['pars'], K)
    ro1, ro2 = liw.compute_weights()

    if kerfPara['type'] == 'lin':
        P = np.dot(np.diag(ro1.flatten()), np.hstack([A, e1]))
        Q = np.dot(np.diag(S2.flatten()), np.hstack([B, e2]))
        H1 = np.dot(P.T, P)
        Q1 = np.dot(Q.T, Q)
        kerH1 = -np.linalg.solve(H1 + c1 * Q1 + epsilon * np.eye(len(Q1)), np.dot(Q.T, np.diag(S2.flatten())) @ e2)
        w1 = kerH1[:-1, :]
        b1 = kerH1[-1, :]

        P = np.dot(np.diag(S1.flatten()), np.hstack([A, e1]))
        Q = np.dot(np.diag(ro2.flatten()), np.hstack([B, e2]))
        H1 = np.dot(P.T, P)
        Q1 = np.dot(Q.T, Q)
        kerH2 = np.linalg.solve(Q1 + c2 * H1 + epsilon * np.eye(len(H1)), np.dot(P.T, np.diag(S1.flatten())) @ e1)
        w2 = kerH2[:-1, :]
        b2 = kerH2[-1, :]

    # Giả sử 'kernelfun' là một hàm bạn đã hoặc cần triển khai
    # Mã cho kernel phi tuyến tương tự như trường hợp tuyến tính

    train_time = time.time() - start_time
    P_1 = TestX
    y1 = (np.dot(P_1, w1) + b1) / np.sqrt(np.dot(w1.T, w1))
    y2 = (np.dot(P_1, w2) + b2) / np.sqrt(np.dot(w2.T, w2))
    Predict_Y = np.sign(np.abs(y2) - np.abs(y1))

    # Dự đoán sử dụng trọng số và sai số tính toán
    # Bước này cần được điều chỉnh dựa trên cách xử lý kernelfun và các tính toán phi tuyến trong Python

    return Predict_Y, train_time, {'function_name': 'IFW_LSTSVM'}
```
